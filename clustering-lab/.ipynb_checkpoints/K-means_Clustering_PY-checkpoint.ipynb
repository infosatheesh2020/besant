```python
#K_means
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
my_cluster = KMeans(n_clusters = 4)
```

```python
import pandas as pd  
import numpy as np
customer_data = pd.read_csv('shopping_data.csv')
customer_data
```

```python
data = customer_data.iloc[:, 3:5].values
```

```python
model = my_cluster.fit(data)
clusters = model.labels_
clusters
```

```python
center_values = model.cluster_centers_
center_values
```

```python
#Plotting centers
plt.scatter(center_values[:, 0],center_values[:, 1], c='black', s=200);
```

```python
#plotting Clusters
#This plots points with cluster dependent colors
plt.figure(figsize=(10, 10))
plt.scatter(customer_data['Annual Income (k$)'], customer_data['Spending Score (1-100)'], c=clusters)  
plt.show()
```
