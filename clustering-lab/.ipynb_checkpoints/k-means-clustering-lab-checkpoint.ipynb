```python
# K- means = python implementation is sklearn package is actually K-means++
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
my_cluster = KMeans(n_clusters=4)
```

```python
import pandas as pd
import numpy as np
customer_data = pd.read_csv('Clustering Shopping_data.csv')
customer_data
```

```python
data = customer_data.iloc[:,3:5]
data
```

```python
help(my_cluster.fit)
```

```python
model = my_cluster.fit(data)
clusters = model.labels_
clusters
```

```python
center_values = model.cluster_centers_
center_values
```

```python
# scatter plot for centroids
plt.scatter(center_values[:,0], center_values[:,1], c='black', s=200)
```

```python
help(plt.scatter)
```

```python
# plot for clusters with different colors
#help(plt.figure)
plt.figure(figsize=(10,10))
plt.scatter(customer_data['Annual Income (k$)'], customer_data['Spending Score (1-100)'], c=clusters)
plt.show()
```
