```python
import matplotlib.pyplot as plt  
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
customer_data = pd.read_csv('Clustering Shopping_data.csv')
customer_data
```

```python
data = customer_data.iloc[:,3:5]
data
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))  
plt.title("Customer Dendograms")  
dend = shc.dendrogram(shc.linkage(data, method='ward')) # linkage is agloromative, ward is for euclidian
```
