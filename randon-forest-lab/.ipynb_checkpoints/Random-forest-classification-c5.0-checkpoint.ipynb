```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Bankruptcy_data.csv")
df.describe()
```

```python
X = df.iloc[:,:-1]
X
```

```python
Y = df.iloc[:,-1:]
Y
```

```python
X.isnull().sum().sum()
```

```python
# perform central imputation on X to remove missing values
from sklearn.preprocessing import Imputer
ci = Imputer()
```

```python
X1 = pd.DataFrame(ci.fit_transform(X))
X1.isnull().sum().sum()
```

```python
# sample only 1000 rows to avoid too much computation
X2 = X1.iloc[:1000, :]
X2
```

```python
Y2 = Y.iloc[:1000, :]
Y2
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
# random forest classifier with 50 trees
rf_clf = RandomForestClassifier(n_estimators=50)
rf_clf
```

```python
# get index of the columns
names = df.dtypes.index
names
```

```python
rf_model = rf_clf.fit(X2, Y2)
rf_model
```

```python
# Get importances of each attribute
rf_model.feature_importances_
```

```python
# Print importances in nice way
print("Features sorted by their score:")
print(sorted(zip(map(lambda x: round(x, 4), rf_model.feature_importances_), names), reverse = True))
```

```python
predictions = pd.DataFrame(rf_model.predict(X2))
confusion_matrix(Y2, predictions)
```
