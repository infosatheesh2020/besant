```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Bankruptcy_data.csv')
data.describe()
```

```python
data.iloc[:50,:]
#data.target.value_counts()
```

# To limit number of rows, find important attributes from the total attributes list
# Random forest classification can be leveraged to identify top 10 important attributes

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import Imputer
```

# Sample 3000 rows and apply random forest algorithm to avoid too much computation

```python
rf_sample_data = data.iloc[:3000,:-1]
rf_data_y = data.iloc[:3000,-1]
rf_sample_data.head()
```

# Perform Imputation for missing values

```python
rf_sample_data.isnull().sum().sum()
```

```python
ci = Imputer()
rf_data = pd.DataFrame(ci.fit_transform(rf_sample_data))
rf_data.isnull().sum().sum()
```

# Build random forest model

```python
rf_clf = RandomForestClassifier(n_estimators=50)
rf_clf
```

```python
names = data.dtypes.index
names
```

```python
rf_model = rf_clf.fit(rf_data,rf_data_y)
rf_model
```

```python
print("Features Sorted by their score")
importance = sorted(zip(map(lambda x : round(x,4), rf_model.feature_importances_), names), reverse=True)
importance
```

# Getting top 10 important attributes and filter data

```python
# Select top 10 attributes with high importance
important_features = [i[1] for i in importance[:][:10]]
important_features
```

```python
x = data[important_features]
x.head()
```
